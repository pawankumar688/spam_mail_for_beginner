```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
import string
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)

    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()


    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)

    text = y[:]
    y.clear()


    for i in text:
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer(max_features=3000)
```

### Block 43 & 44:
        In which TfidfVectorizer() function is a tool that helps computers understand the meaning of words in text. It looks at how often words appear in a piece of text (like an article or a book) and how important those words are in the whole collection of texts. This helps computers figure out which words are most relevant to understand the meaning of a particular text, so we will transform our transformed_text to get the value we desire.

```python
x = tfid.fit_transform(df['transformed_text']).toarray()
```

```python
x.shape
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x.shape, x_train.shape, x_test.shape)
```

### Block 44 to 48:
        This is the step where our traing testing going to start so for the we need other function called train_test_split we will provide parameters to it (x and y, set testing size 0.2 means 20% of you data will for testing and other 80% will be for traning data, provide random state any value you like) check shape for your testing and traing size with the size of original data.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
model_1 = LogisticRegression()
model_1.fit(x_train, y_train)
```

```python
model_2 = RandomForestClassifier(n_estimators=50, random_state=2)
model_2.fit(x_train, y_train)
```

```python
model_3 = SVC(kernel='linear')
model_3.fit(x_train, y_train)
```

### Block 49 to 52:
        Now we are going to import the algorithms for we are going to train our model on so I have chose 3 of them for checking which algorithm provides the better performance over all, so I'm going to use LogesticRegression(), RandomForestClassifier() in which I provide parameters to it like (n_estimater=50, random_state=2 ( this are not the fix values you can change them according to your problem )), and last but not the list SVC() in which we use the kernel='liner'. there are also other algorithm which I haven't mentiond yet but you can study them seprately and I also sugest you, Please do so to better understand the workflow of what are we doing here, it will take time but it will be worth it.

```python
from sklearn.metrics import accuracy_score
```

```python
prediction_train_1 = model_1.predict(x_train)
accuracy_train_1 = accuracy_score(prediction_train_1, y_train)
prediction_test_1 = model_1.predict(x_test)
accuracy_test_1 = accuracy_score(prediction_test_1, y_test)
print('Accuracy for traning 1: ', accuracy_train_1)
print('Accuracy for testing 1: ', accuracy_test_1)
```

```python
prediction_train_2 = model_2.predict(x_train)
accuracy_train_2 = accuracy_score(prediction_train_2, y_train)
prediction_test_2 = model_2.predict(x_test)
accuracy_test_2 = accuracy_score(prediction_test_2, y_test)
print('Accuracy for traning 2: ', accuracy_train_2)
print('Accuracy for testing 2: ', accuracy_test_2)
```

```python
prediction_train_3 = model_3.predict(x_train)
accuracy_train_3 = accuracy_score(prediction_train_3, y_train)
prediction_test_3 = model_3.predict(x_test)
accuracy_test_3 = accuracy_score(prediction_test_3, y_test)
print('Accuracy for traning 3: ', accuracy_train_3)
print('Accuracy for testing 3: ', accuracy_test_3)
```
